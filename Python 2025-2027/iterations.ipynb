```python
#write a program to print even numbers from 1 to 20 using for loop
for i in range(1,21):
    if (i%2==0):
        print(i,end=",")
```

```python
#write a program to print odd numbers from 50 to 100 using for loop
for i in range(50,101):
    if (i%2!=0):
        print(i,end=",")
```

```python
#write a program to find the vowels from a string using for loop
a=str(input("Enter a string: "))
v="aeiou"
for i in a:
    if (i in v):
        print(i,end=",")
    
```

```python
#write a program to find the non vowels from a string using for loop
a=str(input("Enter a string: "))
v="aeiouAEIOU"
for i in a:
    if (i not in v):
        print(i,end=",")
```

```python
#write a program to find the sum of all numbers in between 1 to 20 using for loop
sum=0
for i in range(1,21):
    sum=sum+i
print("The sum of all numbers from 1 to 20 is:",sum)
```

```python
#write a program to find the sum of all odd numbers in between 1 to 1000 using for loop
sum=0
for i in range(1,1001):
    if (i%2!=0):
        sum=sum+i
print("The sum of all odd numbers from 1 to 1000 is:",sum)
```

```python
#write a program to find the product of all even numbers in between 500 to 600 using for loop
p=1
for i in range(500,601):
    if (i%2==0):
        p=p*i
print("The product of all even numbers from 500 to 600 is:",p)
```
