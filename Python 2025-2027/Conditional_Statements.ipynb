#Conditional Statements in python
Python has 3 types of conditional statements
1.If statement 
2.elif statement
3.else statement
```python
if(logic):
  statment
```

```python if, else
if(logic):
    statement
else:
    statement
```

```python if,elif,else
if(logic):
    statements
elif(logic):
    statments
else:
    statement
```


````python
#write a program to check if a number is even or odd
num = int(input("Enter a number: "))
if num % 2 == 0:
    print(f"{num} is even")
else:
    print(f"{num} is odd")
````

````python

#write a code to check if a number is positive, negative or zero
num = float(input("Enter a number: "))
if num > 0:
    print("Positive number")
elif num < 0:
    print("Negative number")
else:
    print("Zero or invalid input")
````

````python
"""
write a program to find the grades of a student based on their marks
1.if marks in between 80-100 then grade is A
2.if marks in between 60-79 then grade is B
3.if marks in between 40-59 then grade is C
4.if marks in between 25-29 then grade is D
5.if marks in between 0-24 then grade is F
6.if marks is less than 0 or greater than 100 then invalid marks
"""
while True:
    m = float(input("Enter your marks (0–100): "))
    
    if 80 <= m <= 100:
        print("Student got Grade A")
        break
    elif 60 <= m <= 79:
        print("Student got Grade B")
        break
    elif 40 <= m <= 59:
        print("Student got Grade C")
        break
    elif 25 <= m <= 39:
        print("Student got Grade D")
        break
    elif 0 <= m <= 24:
        print("Student got Grade F")
        break
    else:
        print("❌ Invalid Marks. Try again.")
````
