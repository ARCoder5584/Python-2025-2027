```python
import math

try:
    import ipywidgets as widgets
    from IPython.display import display, HTML, clear_output
except ImportError:
    raise ImportError("ipywidgets is required for this notebook app.")

def triangle_summary(a, b, c):
    sides = (a, b, c)
    if any(s <= 0 for s in sides):
        return "<b>Error:</b> All side lengths must be positive."

    if a + b <= c or a + c <= b or b + c <= a:
        return "<b>Error:</b> The given sides do not form a triangle."

    perimeter = a + b + c
    s = perimeter / 2
    area = math.sqrt(max(s * (s - a) * (s - b) * (s - c), 0.0))

    side_type = (
        "Equilateral" if a == b == c
        else "Isosceles" if a == b or a == c or b == c
        else "Scalene"
    )

    def angle(opposite, x, y):
        value = (x*x + y*y - opposite*opposite) / (2 * x * y)
        value = max(-1.0, min(1.0, value))
        return math.degrees(math.acos(value))

    A = angle(a, b, c)
    B = angle(b, a, c)
    C = angle(c, a, b)

    if any(abs(x - 90) < 1e-9 for x in (A, B, C)):
        angle_type = "Right"
    elif any(x > 90 for x in (A, B, C)):
        angle_type = "Obtuse"
    else:
        angle_type = "Acute"

    return f"""
    <h3>Triangle Calculator</h3>
    <b>Type:</b> {side_type}<br>
    <b>Angle Class:</b> {angle_type}<br>
    <b>Perimeter:</b> {perimeter:.4f}<br>
    <b>Area:</b> {area:.4f}<br>
    <b>Angles:</b> A={A:.2f}°, B={B:.2f}°, C={C:.2f}°
    """

a = widgets.FloatText(value=3.0, description="Side a")
b = widgets.FloatText(value=4.0, description="Side b")
c = widgets.FloatText(value=5.0, description="Side c")
out = widgets.Output()

def update(*args):
    with out:
        clear_output(wait=True)
        display(HTML(triangle_summary(a.value, b.value, c.value)))

for w in (a, b, c):
    w.observe(update, names="value")

display(widgets.VBox([a, b, c, out]))
# Show the initial triangle summary
update()
```
