```python
# Data Structures in Python
1.Data Structures is collections of data in different formats.
2.5 types of Data Structures in Python:
    a.String-String is a collection of data in single or double quotes '' or "".
    b.List-list is a collection of elements enclosed in square brackets [].
    c.Tuple-Tuple is a collection of elements enclosed in parentheses ().
    d.Dictionary- Dictionary is a collection of elements enclosed in curly brackets {}.
    e.Set-Set is a collection of elements enclosed in curly braces {}.
    
```

```python
#String
s='I will Heck You Bitches'
p="I will Heck You Gays"
print(s)
print(p)
print(type(s))
print(type(p))
```

```python
#List
names=['Harry','Pablo','Beluga','Kitler']
ages=[21,22,23,43]
salaries=[25000,30000,40000,5000000]
print(names)
print(ages)
print(salaries)
print(type(names))
print(type(ages))
print(type(salaries))
```

```python
#Tuple
names=['Elon','Ho Chi mi','Stalin','Hecker']
ages=[21,22,23,13]
salaries=[2,3000000,400000,10000000]
print(names)
print(ages)
print(salaries)
print(type(names))
print(type(ages))
print(type(salaries))
```

```python
#Dictionary
Germany={'Germany':'Hitler','Aaryan Army':2000000,'Capital':'Berlin'}
Russia={'Russia':'Stalin','Red Army':3000000,'Capital':'Moscow'}
print(Germany)
print(Russia)
print(type(Germany))
print(type(Russia))
```

```python
#Set
States={'Delhi',' Berlin','Moscow','New York'}
Countries={'India','Germany','Russia','USA'}
print(States)
print(Countries)
print(type(States))
print(type(Countries))
```

```python
#Strings
s1 = "Hello, World!"
s2='Python is fun.'
s3='''$sudo heck 10923 
$sudo Launch Nuke~10923
if ^successful then^~ die mfs
else~Hecker will be beeck.'''
print(s1)
print(s2)
print(s3)
print(type(s1))
print(type(s2))
print(type(s3))
```

```python
#Concatenation of Strings
greeting = "Heil"
name = "Kitler"
print(greeting + ", " + name + "!") 
```

```python
#String repetition
Command = "sudo heck \n"
print(Command * 31000) 
```

```python
#accessing characters in a string
#2 ways: 1.Forward Indexing 2.Backward Index
#forward index-left to right()
#backward index -right to left ()
str1 = "Heil Kitler"
print(str1[0])  #H
print(str1[5])  #K
print(str1[-1]) #r
print(str1[-6]) #K
```
