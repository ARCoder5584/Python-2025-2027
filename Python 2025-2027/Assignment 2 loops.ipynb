```python
a=int(input("Enter a non-negative integer: "))

if (a<0):
    print("Factorial is not defined for negative numbers.")
elif (a==0):
    print("The factorial of 0 is 1.")
else:
    fa = 1
    for i in range(1, a + 1):
        fa *= i
    print("The factorial of number is.",fa)
```
